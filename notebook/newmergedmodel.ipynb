### merging dataset 1 and 2

```python
import pandas as pd

# 1. Load both datasets
df1 = pd.read_csv("/Users/sayenaqureshi/Main-Folder/datasets/Crop and fertilizer dataset (2).csv")  # has Crop
df2 = pd.read_csv("/Users/sayenaqureshi/Main-Folder/datasets/Crop_recommendation (3).csv")  # has label

# 2. Rename dataset2 columns to match dataset1
rename_map = {
    "N": "Nitrogen",
    "P": "Phosphorus",
    "K": "Potassium",
    "temperature": "Temperature",
    "ph": "pH",
    "rainfall": "Rainfall",
    "label": "Crop"   # <-- IMPORTANT
}
df2 = df2.rename(columns=rename_map)

# 3. Add missing columns to df2 (undefined/NaN)
cols_only_in_df1 = ["District_Name", "Soil_color", "Fertilizer", "Link"]
for col in cols_only_in_df1:
    if col not in df2.columns:
        df2[col] = pd.NA

# 4. Add missing column to df1 (undefined/NaN)
if "humidity" not in df1.columns:
    df1["humidity"] = pd.NA

# 5. final column order
final_cols = [
    "District_Name",
    "Soil_color",
    "Nitrogen",
    "Phosphorus",
    "Potassium",
    "pH",
    "Rainfall",
    "Temperature",
    "humidity",
    "Crop",
    "Fertilizer",
    "Link"
]

df1 = df1[final_cols]
df2 = df2[final_cols]

# 6. Merge (vertical stack)
merged_df = pd.concat([df1, df2], ignore_index=True)

# 7. Save
merged_df.to_csv("merged_dataset.csv", index=False)
```
